### Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

**Answer:**  
Anomaly Detection is the process of identifying unusual patterns, data points, or events that do not follow the normal behavior of a dataset. These unusual points are called anomalies or outliers, and they often indicate critical incidents such as fraud, network intrusions, system failures, or abnormal trends.  

There are mainly three types of anomalies:  

---

#### 1. Point Anomalies  
- A single data point that is very different from the rest of the data.  
- **Example:** In a dataset of monthly expenses where most values are around ₹10,000–₹20,000, an expense of ₹1,00,000 in one month is a point anomaly.  

---

#### 2. Contextual Anomalies  
- A data point that is unusual in a specific context but may be normal in another.  
- **Example:** A temperature of 30°C is normal in summer but would be an anomaly in winter.  

---

#### 3. Collective Anomalies  
- A group of data points that are normal individually but abnormal when considered together.  
- **Example:** In a network traffic log, if thousands of requests come from the same IP within a few seconds, each request is normal, but collectively it indicates a possible cyberattack.  

---

**In summary:**  
Anomaly detection helps in finding rare, suspicious, or unexpected patterns in data, which are very important for security, fraud detection, fault diagnosis, and many real-life applications.  

### Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

**Answer:**  
Anomaly detection can be done using different algorithms depending on the type of data and problem. Three popular methods are **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)**.

---

#### 1. Isolation Forest (IF)  
**Approach:**  
- Based on the idea that anomalies are “few and different.”  
- Uses random decision trees to isolate data points.  
- Anomalies are isolated faster because they need fewer splits.  

**Best For:**  
- High-dimensional data.  
- Large datasets.  
- Works well in fraud detection, system monitoring, etc.  

---

#### 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)  
**Approach:**  
- Groups data points into clusters based on density.  
- Points that do not belong to any cluster (too far away) are treated as anomalies.  

**Best For:**  
- Data with clusters of different shapes.  
- Spatial/geographical data.  
- Example: detecting unusual locations in GPS data or abnormal traffic patterns.  

---

#### 3. Local Outlier Factor (LOF)  
**Approach:**  
- Compares the density of a point with the density of its neighbors.  
- A point is an outlier if it is in a region of lower density compared to its neighbors.  

**Best For:**  
- Data with varying local densities.  
- Detecting subtle anomalies that depend on neighborhood relations.  
- Example: detecting unusual customer behavior in e-commerce.  

---

###  Comparison Table

| **Method**        | **Approach**                                  | **Suitable For**                    |
|--------------------|-----------------------------------------------|-------------------------------------|
| Isolation Forest   | Random tree splits isolate anomalies quickly | Large, high-dimensional datasets    |
| DBSCAN             | Density-based clustering, anomalies = noise  | Spatial/clustered data              |
| LOF                | Density comparison with neighbors            | Local density variations            |

---

**Conclusion:**  
- **Isolation Forest** → Best for large, high-dimensional data.  
- **DBSCAN** → Best when data has clusters and noise.  
- **LOF** → Best when anomalies depend on local density differences.  
 

### Question 3: What are the key components of a Time Series? Explain each with one example.  

**Answer:**  
A **Time Series** is a sequence of data points collected over time, usually at equal intervals (daily, monthly, yearly, etc.).  
To properly analyze time series, we break it into four key components:  

---

#### 1. Trend  
- **Definition:** Long-term movement or direction in data over a long period (upward or downward).  
- **Example:** The steady increase in the number of internet users worldwide over the last 20 years.  

---

#### 2. Seasonality  
- **Definition:** Regular and repeating patterns within a fixed time period (daily, monthly, yearly).  
- **Example:** Ice cream sales increase every summer and decrease in winter.  

---

#### 3. Cyclic Component  
- **Definition:** Fluctuations that occur over a longer time frame (more than a year), often related to economic or business cycles. Unlike seasonality, cycles are not fixed or regular.  
- **Example:** Rise and fall of stock markets depending on economic conditions.  

---

#### 4. Irregular (Random) Component  
- **Definition:** Unpredictable, irregular variations caused by unexpected events.  
- **Example:** Sudden drop in airline travel due to the COVID-19 pandemic.  

---

###  Summary Table  

| Component   | Meaning                              | Example                          |  
|-------------|--------------------------------------|----------------------------------|  
| **Trend**   | Long-term increase/decrease          | Internet users rising over years |  
| **Seasonality** | Repeats in a fixed cycle (short-term) | Summer ice cream sales          |  
| **Cyclic**  | Long-term economic/business cycles   | Stock market ups and downs       |  
| **Irregular** | Random/unexpected fluctuations     | COVID-19 impact on travel        |  

---

**Conclusion:**  
The four components — **Trend, Seasonality, Cyclic, and Irregular** — help us understand patterns in time series data.  
By separating them, we can make better predictions and decisions in fields like economics, business, weather forecasting, and healthcare.  

## Question 4  
**Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**  

### Answer:  
In time series analysis, a dataset is called **stationary** if its statistical properties (**mean, variance, covariance**) remain constant over time.  
Stationarity is important because most forecasting models (like **ARIMA**) work properly only with stationary data.  

---

### 1. Stationary Series  
- **Definition:** A stationary series has no trend, no changing variance, and no seasonal effects.  
- **Example:** Daily temperature deviations from the average (with no overall trend).  

---

### 2. Non-Stationary Series  
- Shows **trend, seasonality, or changing variance** over time.  
- **Example:** Monthly sales data that continuously increase due to business growth.  

---

### 3. How to Test Stationarity  
- **Visual Inspection:** Plot the data and check if mean/variance change over time.  
- **Statistical Tests:**  
  - **ADF (Augmented Dickey-Fuller) Test:** Null hypothesis = data is non-stationary.  
  - **KPSS Test:** Null hypothesis = data is stationary.  

---

### 4. How to Transform Non-Stationary → Stationary  
- **Differencing:** Subtract current value from the previous value to remove trend.  
  - Example: `Yt – Yt-1`  
- **Log Transformation:** Reduce variance in data with large fluctuations.  
  - Example: `log(sales)`  
- **Detrending:** Remove the estimated trend (linear or polynomial).  
- **Seasonal Adjustment:** Remove repeating seasonal patterns using decomposition.  

---

### Conclusion  
A **stationary time series** is essential for reliable forecasting.  
We can detect non-stationarity using **plots and statistical tests (ADF/KPSS)** and make the series stationary by applying **differencing, log transform, detrending, or seasonal adjustment**.

### Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.  

**Answer:**  
Time series forecasting uses different models depending on the data’s properties (trend, seasonality, external factors).  
The main models are **AR, MA, ARIMA, SARIMA, and SARIMAX**.  

---

#### 1. Autoregressive (AR) Model  
- **Structure:** Uses past values of the series to predict the future (depends on lagged observations).  
- **Equation:**  
  \[
  Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + ... + \epsilon_t
  \]  
- **Application:** Works well when the current value is strongly correlated with previous values (e.g., stock prices, weather data).  

---

#### 2. Moving Average (MA) Model  
- **Structure:** Uses past forecast errors (residuals) to predict future values.  
- **Equation:**  
  \[
  Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + ... + \epsilon_t
  \]  
- **Application:** Useful when random shocks/errors have short-term influence (e.g., noise in sensor readings).  

---

#### 3. ARIMA (Autoregressive Integrated Moving Average)  
- **Structure:** Combines AR and MA, with an additional “I” (Integration) step to make data stationary.  
- **Notation:** ARIMA(p, d, q)  
  - *p* = AR terms  
  - *d* = differencing (to remove trend)  
  - *q* = MA terms  
- **Application:** General-purpose forecasting model for time series with **trend but no strong seasonality** (e.g., GDP growth, sales trends).  

---

#### 4. SARIMA (Seasonal ARIMA)  
- **Structure:** Extends ARIMA by adding seasonal components.  
- **Notation:** SARIMA(p, d, q)(P, D, Q, s)  
  - (P, D, Q) = seasonal AR, differencing, MA  
  - *s* = seasonal period (e.g., 12 for monthly data)  
- **Application:** Best for data with **both trend and seasonality** (e.g., airline passenger data, electricity demand by season).  

---

#### 5. SARIMAX (Seasonal ARIMA with Exogenous Variables)  
- **Structure:** Same as SARIMA but includes external (exogenous) factors.  
- **Application:** Used when **external variables** influence the series.  
- **Example:** Forecasting sales using seasonal trends + external features like marketing spend or holidays.  

---

###  Comparison Table  

| Model   | Structure                  | Handles Trend | Handles Seasonality | Uses External Factors | Application                  |  
|---------|----------------------------|---------------|----------------------|------------------------|------------------------------|  
| **AR**  | Past values                | No            | No                   | No                     | Stock prices                 |  
| **MA**  | Past errors                | No            | No                   | No                     | Sensor noise                 |  
| **ARIMA** | AR + Differencing + MA   | Yes           | No                   | No                     | GDP, sales trends            |  
| **SARIMA** | ARIMA + Seasonal terms  | Yes           | Yes                  | No                     | Airline passengers           |  
| **SARIMAX** | SARIMA + External vars | Yes           | Yes                  | Yes                    | Sales with promotions        |  

---

**Conclusion:**  
- **AR & MA** are simple base models.  
- **ARIMA** is effective for **trend-based series**.  
- **SARIMA** handles both **trend and seasonality**.  
- **SARIMAX** is the most advanced, capturing **trend, seasonality, and external factors** for highly accurate forecasting.

### Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components

```python
#Answer
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("AirPassengers.csv")
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Plot original series
plt.figure(figsize=(10,5))
plt.plot(data['#Passengers'], label="Original Series")
plt.title("AirPassengers Dataset - Original Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Calculate moving averages (Trend estimation)
data['Trend'] = data['#Passengers'].rolling(window=12, center=True).mean()

# Detrended data (Seasonal + Residual)
data['Detrended'] = data['#Passengers'] - data['Trend']

# Plot Trend
plt.figure(figsize=(10,5))
plt.plot(data['Trend'], color='red', label="Trend (12-Month Moving Avg)")
plt.title("Trend Component")
plt.legend()
plt.show()

# Plot Detrended (Seasonal + Residual)
plt.figure(figsize=(10,5))
plt.plot(data['Detrended'], color='green', label="Seasonality + Residuals")
plt.title("Seasonality + Residual Component (Approx.)")
plt.legend()
plt.show()
```

### Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot. 

```python
#Answer
# -----------------------------
# Isolation Forest on NYC Taxi Fare Dataset
# -----------------------------

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# 1. Load dataset safely
data = pd.read_csv("NYC_taxi_fare_data.csv", low_memory=False)

# Handle common mixed-type columns (convert safely)
if 'pickup_datetime' in data.columns:
    data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'], errors='coerce')
if 'passenger_count' in data.columns:
    data['passenger_count'] = pd.to_numeric(data['passenger_count'], errors='coerce')

# 2. Select numeric features for anomaly detection
X = data[['fare_amount', 'trip_distance']].dropna()

# 3. Apply Isolation Forest
iso = IsolationForest(contamination=0.02, random_state=42)
data = data.loc[X.index]   # keep only rows with no NaN in selected features
data['anomaly'] = iso.fit_predict(X)

# Separate anomalies and normal points
anomalies = data[data['anomaly'] == -1]
normals = data[data['anomaly'] == 1]

# 4. Sample data for faster + clear plotting
sampled_normals = normals.sample(n=5000, random_state=42) if len(normals) > 5000 else normals
sampled_anomalies = anomalies.sample(n=500, random_state=42) if len(anomalies) > 500 else anomalies

# 5. Plot anomalies vs normal points
plt.figure(figsize=(10,6))
plt.scatter(sampled_normals['trip_distance'], sampled_normals['fare_amount'],
            c='blue', label="Normal", alpha=0.4, s=10)
plt.scatter(sampled_anomalies['trip_distance'], sampled_anomalies['fare_amount'],
            c='red', label="Anomaly", alpha=0.7, s=20)
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest - Anomaly Detection (NYC Taxi Fare)")
plt.legend(loc="upper right")
plt.show()
```

### Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results. 


```python
#Answer
#!pip install statsmodels 


# -----------------------------
# SARIMA Forecasting on AirPassengers Dataset
# -----------------------------

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 1. Load dataset
data = pd.read_csv("AirPassengers.csv")

# 2. Convert 'Month' to datetime and set index with monthly frequency
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
data = data.asfreq('MS')   # avoid frequency warning

# 3. Define target series
y = data['#Passengers']

# 4. Train SARIMA model
model = SARIMAX(y, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# 5. Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# 6. Plot original data + forecast
plt.figure(figsize=(12,6))
plt.plot(y, label="Observed")
plt.plot(forecast.predicted_mean, label="Forecast", color='red')

# Confidence intervals
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1],
                 color='pink', alpha=0.3)

plt.title("SARIMA Forecast - AirPassengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

### Question 10: You are working as a data scientist for a power grid monitoring company.

Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features

```python
#Answer
# - 15-min data
# - Isolation Forest for anomalies (fast)
# - SARIMAX(…, 96) for daily seasonality
# - Walk-forward style split, metrics, plots
# ============================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ---------------------------
# 0) Get data (demo = synthetic)
#    In production, replace this block with:
#    df = pd.read_csv("your_file.csv", parse_dates=["timestamp"])
#    df = df.sort_values(["region","timestamp"])
# ---------------------------
np.random.seed(7)
idx = pd.date_range("2024-01-01", periods=96*7, freq="15min")  # 7 days of 15-min data
base = 300 + 40*np.sin(np.linspace(0, 14*np.pi, len(idx)))     # diurnal-ish pattern
noise = np.random.normal(0, 8, len(idx))
energy = base + noise

# inject a few anomalies
energy[320:324] += 120   # spike
energy[515:518] -= 90    # drop

df = pd.DataFrame({
    "timestamp": idx,
    "region": "R1",
    "temp_c": 20 + 5*np.sin(np.linspace(0, 4*np.pi, len(idx))), # simple weather proxy
    "energy_usage": energy
})
df = df.set_index("timestamp").asfreq("15min")

# ---------------------------
# 1) Real-time anomaly detection (rolling window)
# ---------------------------
# For streaming, use the last N hours; here we use entire series for a quick demo.
# If you have features (temp, region), include them too.
anomaly_features = ["energy_usage"]  # extend with exogenous if desired

iso = IsolationForest(
    n_estimators=150,
    contamination=0.02,     # ~2% anomalies; tune to your grid
    random_state=42
)
df["anomaly"] = iso.fit_predict(df[anomaly_features])  # -1 = anomaly, 1 = normal

# ---------------------------
# 2) Short-term forecasting with SARIMAX
# ---------------------------
# Seasonal period = 96 (daily seasonality at 15-min frequency)
# Use exogenous weather (temp_c) as an example
y = df["energy_usage"]
X = df[["temp_c"]]

# Train/test split: last 24 steps (6 hours) as test
h = 24
y_train, y_test = y.iloc[:-h], y.iloc[-h:]
X_train, X_test = X.iloc[:-h], X.iloc[-h:]

# Light-weight SARIMAX spec to run fast; tune for production
model = SARIMAX(
    y_train,
    exog=X_train,
    order=(1,0,1),              # small ARMA core (no differencing for speed; add if needed)
    seasonal_order=(1,1,1,96),  # daily seasonality; seasonal differencing helps stationarity
    enforce_stationarity=False,
    enforce_invertibility=False
)
res = model.fit(disp=False)

fc = res.get_forecast(steps=h, exog=X_test)
fc_mean = fc.predicted_mean
fc_ci = fc.conf_int()

# ---------------------------
# 3) Validation metrics
# ---------------------------
rmse = np.sqrt(mean_squared_error(y_test, fc_mean))
mape = mean_absolute_percentage_error(y_test, fc_mean) * 100

print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")

# ---------------------------
# 4) Plots: anomalies + forecast
# ---------------------------
# (a) Anomaly view
plt.figure(figsize=(12,5))
plt.plot(df.index, df["energy_usage"], label="Energy usage")
plt.scatter(df.index[df["anomaly"] == -1],
            df.loc[df["anomaly"] == -1, "energy_usage"],
            s=25, color="red", label="Anomaly")
plt.title("Anomaly Detection (Isolation Forest)")
plt.xlabel("Time")
plt.ylabel("kWh")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()

# (b) Forecast view
plt.figure(figsize=(12,5))
plt.plot(y_train.index, y_train, label="Train")
plt.plot(y_test.index, y_test, label="Actual (Test)", linewidth=2)
plt.plot(fc_mean.index, fc_mean, label="Forecast (Next 6 hours)", linewidth=2)

# confidence interval
plt.fill_between(fc_ci.index, fc_ci.iloc[:,0], fc_ci.iloc[:,1], alpha=0.2)

plt.title(f"SARIMAX Forecast (h={h})  |  RMSE={rmse:.1f}, MAPE={mape:.1f}%")
plt.xlabel("Time")
plt.ylabel("kWh")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()

# ---------------------------
# 5) (Optional) Streaming pattern
# ---------------------------
# Pseudocode idea:
# for each new 15-min batch:
#   append batch -> keep last N days window
#   score anomalies with fitted IF (or refit on rolling window periodically)
#   update SARIMAX with new observation (or refit nightly); produce next-h forecast
#   log errors; if rolling MAPE > threshold -> trigger retrain/alert
```
